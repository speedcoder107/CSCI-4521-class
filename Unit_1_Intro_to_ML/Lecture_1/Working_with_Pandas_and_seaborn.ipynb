Working with Pandas and seaborn

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
wage_df = pd.read_csv('https://raw.githubusercontent.com/sziccardi/CSCI4521_DataRepository/main/Wage.csv')

print(wage_df.columns)
```

```python
print(wage_df.describe)
```

```python
plt.plot(wage_df.age, wage_df.wage, 'o')
plt.xlabel('Age')
plt.ylabel('Wage')
```

```python
wage_df.maritl.value_counts()
```

```python

married_df = wage_df[wage_df.maritl == '2. Married']
print(married_df)
# married_df.shape
```

'hue' is used to seperate data via a third dimension

```python
plt.figure()
sns.histplot(data=wage_df, x = 'wage', hue='jobclass', kde='True')
plt.xlabel('Wage')
plt.ylabel('Frequency')
```

This creates pair plots

```python
sns.pairplot(wage_df, vars=['year', 'age', 'wage'], hue='jobclass')
```
