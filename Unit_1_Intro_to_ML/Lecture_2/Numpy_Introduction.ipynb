```python
# Create a matrix with nested lists
x = [[1,2],[3, 4]]
print(x)
```

```python
# Numpy can do this too!
x = np.array([[1,2],[3, 4]])
print(x)
```

```python
# Reshape command lets you change shape (e.g., turn list to matrix)
print("Original:")
x = np.array([1, 2, 3, 4])
print(x)
print("Reshaped:")
x = np.reshape(x, [2,2])
print(x)
```

```python
#Numpy supports many built-in matrix operations
s = np.sqrt(x)
print(s)
```

```python
s*s
```

```python
# Creating new arrays from old ones is supirsingly useful
A = np.array([1,1,2,3,5,8])
print(A)
print(np.ones(A.size))
print(np.zeros(A.size))
```

## Plotting in Numpy

```python
import matplotlib.pyplot as plt
```

```python
xx, yy = np.meshgrid(np.arange(-4, 4, 0.01), np.arange(-4, 4, 0.01))
z = np.cos(xx)*np.cos(yy)
plt.contour(xx,yy,z)
```

```python
yy
```

```python
plt.pcolormesh(xx, yy, z)
```

```python
# imshow looks a bit prettier, extent can specify axis bounds
plt.imshow(z, extent=(xx[-1,0], xx[0,-1], yy[-1,0], yy[0,0]) )
```

```python
from mpl_toolkits.mplot3d import axes3d
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(xx, yy, z)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(xx, yy, z, cmap='viridis', edgecolor='none')
ax.set_title("sin(x)*cos(y)")
plt.show()
```
